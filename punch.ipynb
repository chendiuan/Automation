```python
!pip install selenium
!apt-get update 
!apt install chromium-chromedriver

from selenium import webdriver
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
import time
import requests

###########################################
account_id = "deantychen"
account_pw = "Ji394su31150"
###########################################

class PunchCard():
    def __init__(self):
        self.main()

    def main(self):
        self.login_vpn()
        self.login_cms()
        #self.punch()
        # self.grab_record()
        self.logout()
        token = '1lnZqdpfwUVv5kMaeMcKAfXB5jAbxa0Szm4IxkE0E5n'
        msg = self.alert_msg
        self.lineNotifyMessage(token, msg)

        # r_msg = self.record
        # self.lineNotifyMessage(token, r_msg)

    def lineNotifyMessage(self, token, msg):

        headers = {
            "Authorization": "Bearer " + token,
            "Content-Type" : "application/x-www-form-urlencoded"
        }

        payload = {'message': msg }
        r = requests.post("https://notify-api.line.me/api/notify", headers = headers, params = payload)
        return r.status_code

    def login_vpn(self):
        #open browser and connect to VPN
        self.driver = webdriver.Chrome('chromedriver',chrome_options=chrome_options)
        self.driver.implicitly_wait(10)
        self.driver.get("https://vpn.ami.com.tw/remote/login?lang=en")

        #input account
        self.driver.find_element("id", "username").send_keys(account_id)
        time.sleep(1)
        self.driver.find_element("id", "credential").send_keys(account_pw)
        time.sleep(1)
        self.driver.find_element("id", "login_button").click()
        time.sleep(5)

        #click login
        self.driver.find_element("xpath", "//body/section[@id='navbar-view-section']/div[1]/div[1]/div[2]/div[2]/div[1]/button[1]").click()
        time.sleep(5)

    def login_cms(self):
        #input cms account
        self.driver.switch_to.window(self.driver.window_handles[1])
        self.driver.find_element("id", "email").send_keys(f"{account_id}@ami.com")
        time.sleep(1)
        self.driver.find_element("id", "password").send_keys(account_pw)
        time.sleep(1)
        self.driver.find_element("id", "login").click()
        time.sleep(5)

    def punch(self):
        #click punch card
        self.driver.find_element("id", "btn_punch_card").click()
        time.sleep(5)
        self.get_alert = self.driver.switch_to.alert
        self.alert_msg = self.get_alert.text
        self.get_alert.accept()
        print (self.alert_msg)
        return self.alert_msg

    def grab_record(self):
        #grab the punch record
        self.driver.find_element("id", "img_time_card_add").click()
        time.sleep(1)
        self.record = self.driver.find_element("xpath", "/html[1]/body[1]/table[1]/tbody[1]/tr[2]/td[1]/table[1]/tbody[1]/tr[1]/td[1]/table[1]/tbody[1]").text
        print (self.record)
        return self.record

    def logout(self):
    #logout accunt
        time.sleep(5)
        self.driver.switch_to.window(self.driver.window_handles[0])
        time.sleep(5)
        self.driver.find_element("xpath", "//header/div[4]/button[3]").click()
        time.sleep(2)
        self.driver.find_element("xpath", "//body/div[3]/div[1]/button[1]").click()
        time.sleep(2)
        #close browser
        self.driver.quit()


if __name__ == "__main__":
    PunchCard()
```
